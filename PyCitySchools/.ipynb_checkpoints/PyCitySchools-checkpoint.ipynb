# PyCity Schools Analysis

Math and Reading scores and passing rates were analyzed to spot any underlying trends. The data comes from nearly 40,000 High School students, distributed among 15 schools. All students are in grades 9 to 12.

Important sidenote: For clarity and ease of understanding, scores are rounded to the nearest tenth, and percentages are rounded to the nearest hundreth, both in analysis and in the DataFrames.

The following trends have been observed:


1. STUDENTS SCORE BETTER IN READING THAN IN MATH.
Overall, students have a much better score in Reading (81.9) than in Math (79.0)


2. SCORES ARE RELATIVELY CONSISTENT BY GRADE.
While average scores are different by school,the scores per Grade Level at each school are rarely off by more than 2 points.


3. SPENDING PER STUDENT IS INVERSELY PROPORTIONAL SCORING.
The more money spent per capita on a student leads to lower scoring in BOTH Math and Reading.


4. CHARTER SCHOOLS OUTPERFORM THEIR DISTRICT COUNTERPARTS, ESPECIALLY IN MATH.
Compared by School Type, Charter School students Score were higher than District students by 7.4 points. They also outperform District Students in Reading, but by less than 3 points. 


# Combined DataSet of Schools and Students

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

- 40,000 High School students 
- 15 schools

- Over $24.6 Million on School funding
- Reading scores (81.9 Avg) higher than math scores (79.0 Avg)

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["school_name"].value_counts())
```

```python
# Calculate the total number of students
student_count = len(school_data_complete["Student ID"].value_counts())
```

```python
# Calculate the total budget
budgetArray = school_data_complete["budget"].unique()
total_budget = budgetArray.sum()
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
```

```python
both_failing_rate = 100 - passing_math_percentage - passing_reading_percentage + overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    "Total Schools": school_count,
    "Total Students": student_count,
    "Total Budget": total_budget,
    "Average Math Score": round(average_math_score, 1),
    "Average Reading Score": round(average_reading_score, 1),
    "% Passing Math": round(passing_math_percentage, 2),
    "% Passing Reading": round(passing_reading_percentage, 2),
    "% Overall Passing": round(overall_passing_rate, 2),
    "% Failing Both": round(both_failing_rate, 2)
}, index = [0])

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
school_name = ["Huang High School",
               "Figueroa High School",
               "Shelton High School",
               "Hernandez High School",
               "Griffin High School",
               "Wilson High School",
               "Cabrera High School",
               "Bailey High School",
               "Holden High School",
               "Pena High School",
               "Wright High School",
               "Rodriguez High School",
               "Johnson High School",
               "Ford High School",
               "Thomas High School",
              ]
```

```python
# Use the code provided to select all of the school types
school_type = [
    "District",
    "District",
    "Charter",
    "District",
    "Charter",
    "Charter",
    "Charter",
    "District",
    "Charter",
    "Charter",
    "Charter",
    "District",
    "District",
    "District",
    "Charter",
]   
```

```python
# Calculate the total student count per school
per_school_counts = []
i = 0
for i in range(school_count):
    kidCount = school_data_complete.loc[school_data_complete["School ID"] == i, :]["student_name"].count()
    per_school_counts.append(kidCount)
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = []
i = 0
for i in range(school_count):
    budgetCount = school_data_complete.loc[school_data_complete["School ID"] == i, :]["budget"].mean()
    per_school_budget.append(budgetCount)

per_school_capita = [x / y for x,y in zip(per_school_budget, per_school_counts)]
```

```python
# Calculate the average test scores per school
per_school_math = []
i = 0
for i in range(school_count):
    mathScore = school_data_complete.loc[school_data_complete["School ID"] == i, :]["math_score"].mean()
    per_school_math.append(mathScore)
per_school_math

per_school_reading = []
j = 0
for j in range(school_count):
    readScore = school_data_complete.loc[school_data_complete["School ID"] == j, :]["reading_score"].mean()
    per_school_reading.append(readScore)
```

```python
per_school_math = [round(x, 1) for x in per_school_math ]
per_school_reading = [round(y, 1) for y in per_school_reading ]
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = []
i = 0
for i in range(school_count):
    schoolTest = school_data_complete.loc[school_data_complete["School ID"] == i, :]
    mathPass = schoolTest[schoolTest["math_score"] >= 70].count()["student_name"]
    students_passing_math.append(mathPass)
    school_students_passing_math = students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = []
i = 0
for i in range(school_count):
    schoolTest = school_data_complete.loc[school_data_complete["School ID"] == i, :]
    readPass = schoolTest[schoolTest["reading_score"] >= 70].count()["student_name"]
    students_passing_reading.append(readPass)
    school_students_passing_reading = students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = []
#i = 0
for i in range(school_count):
    schoolTest = school_data_complete.loc[school_data_complete["School ID"] == i, :]
    bothPass = ( (schoolTest[schoolTest["math_score"] >= 70].count()["student_name"]) & (schoolTest[schoolTest["reading_score"] >= 70].count()["student_name"]) )
    students_passing_math_and_reading.append(bothPass)
    school_students_passing_math_and_reading = students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = [ round(x / y * 100, 2) for x,y in zip(school_students_passing_math, per_school_counts)]
per_school_passing_reading = [ round(x / y * 100, 2) for x, y in zip(school_students_passing_reading, per_school_counts)]
overall_passing_rate = [ round((x / y) * 100, 2) for x,y in zip(school_students_passing_math_and_reading, per_school_counts)]
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    " ": school_name,
    "School Type": school_type,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Total Students": per_school_counts,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_passing_math,
    "% Passing Reading": per_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})

per_school_summary = per_school_summary.set_index(" ")

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]


# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grader_math_scores = []
for i in range(school_count):
    math9Score = ninth_graders.loc[ninth_graders["School ID"] == i, :]["math_score"].mean()
    ninth_grader_math_scores.append(math9Score)

tenth_grader_math_scores = []
for i in range(school_count):
    math10Score = tenth_graders.loc[tenth_graders["School ID"] == i, :]["math_score"].mean()
    tenth_grader_math_scores.append(math10Score)

eleventh_grader_math_scores = []
for i in range(school_count):
    math11Score = eleventh_graders.loc[eleventh_graders["School ID"] == i, :]["math_score"].mean()
    eleventh_grader_math_scores.append(math11Score)

twelfth_grader_math_scores = []
for i in range(school_count):
    math12Score = twelfth_graders.loc[twelfth_graders["School ID"] == i, :]["math_score"].mean()
    twelfth_grader_math_scores.append(math12Score)

    
ninth_grader_math_scores = [round(w, 1) for w in ninth_grader_math_scores]
tenth_grader_math_scores = [round(x, 1) for x in tenth_grader_math_scores]
eleventh_grader_math_scores = [round(y, 1) for y in eleventh_grader_math_scores]
twelfth_grader_math_scores = [round(z, 1) for z in twelfth_grader_math_scores]

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    " ": school_name,
    "9th": ninth_grader_math_scores,
    "10th": tenth_grader_math_scores,
    "11th": eleventh_grader_math_scores,
    "12th": twelfth_grader_math_scores
})

# Remodel into correct format
math_scores_by_grade = math_scores_by_grade.set_index(" ")

# Display the DataFrame
math_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels = spending_labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = round( school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean(), 1)

spending_reading_scores = round( school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean(), 1)

spending_passing_math = round( school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean(), 2)

spending_passing_reading = round( school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean(), 2)

overall_passing_spending = round( school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean(), 2)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Spending Ranges (Per Student)": spending_labels,
    "Average Math Score": spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending
})

spending_summary = spending_summary.set_index("Spending Ranges (Per Student)")

# Display results
spending_summary
```
